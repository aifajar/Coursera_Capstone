Import libraries

```python
import pandas as pd
```

Load dataset from part 1 and geospatial dataset.

```python
df1 = pd.read_csv('neighborhood_in_Toronto.csv')
df1.head()
```

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs_v1/Geospatial_Coordinates.csv'

df2 = pd.read_csv(url)
df2.head()
```

Rename columns

```python
df2.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df2.head()
```

Merge dataframes df1 and df2.

```python
df_toronto = pd.merge(df1, df2, on='PostalCode')

print(df_toronto.shape)
df_toronto.head()
```

```python
df_toronto.to_csv('geo_coordinates_neighborhood_in_Toronto.csv', index=False)
```
